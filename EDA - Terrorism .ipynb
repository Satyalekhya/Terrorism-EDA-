# Terrorsim - EDA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 1: Load the dataset into a Pandas DataFrame with the appropriate encoding
df = pd.read_csv('terrorism_data.csv', encoding='latin-1')
df.head()
```

```python
# Drop columns with all NaN values
df2 = df.dropna(axis=1, how='all')

# Print the filtered DataFrame
print("Filter DataFrame:")
print(df2.head())
```

```python
df2.head()
```

# EDA

```python
# Step -1 :  Print the summary statistics for the filtered columns
summary_stats_filtered = df2.describe()
print("\nSummary Statistics for Filtered Columns:")
print(summary_stats_filtered)
```

```python
# Step 2: Event Count by Year
event_count_by_year = df2['iyear'].value_counts().sort_index()

# Plotting the event count by year
plt.figure(figsize=(12, 6))
plt.plot(event_count_by_year)
plt.xlabel('Year')
plt.ylabel('Number of Events')
plt.title('Number of Terrorism Events by Year')
plt.grid(True)
plt.show()
```

```python
# Step 3: Event Count by Region and Country
event_count_by_region = df2['region_txt'].value_counts()
event_count_by_country = df2['country_txt'].value_counts().head(10)  # Top 10 countries

# Plotting the event count by region
plt.figure(figsize=(12, 6))
event_count_by_region.plot(kind='bar')
plt.xlabel('Region')
plt.ylabel('Number of Events')
plt.title('Number of Terrorism Events by Region')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Step 3  Plotting the event count by country (top 10)
plt.figure(figsize=(12, 6))
event_count_by_country.plot(kind='bar')
plt.xlabel('Country')
plt.ylabel('Number of Events')
plt.title('Number of Terrorism Events by Country (Top 10)')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Step 4: Event Type Distribution
event_type_distribution = df2['attacktype1_txt'].value_counts()

# Plotting the event type distribution
plt.figure(figsize=(12, 6))
event_type_distribution.plot(kind='bar')
plt.xlabel('Event Type')
plt.ylabel('Number of Events')
plt.title('Distribution of Terrorism Event Types')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Step 5: Target Type Distribution
target_type_distribution = df2['targtype1_txt'].value_counts()

# Plotting the target type distribution
plt.figure(figsize=(12, 6))
target_type_distribution.plot(kind='bar')
plt.xlabel('Target Type')
plt.ylabel('Number of Events')
plt.title('Distribution of Terrorism Target Types')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Step 6: Casualty Analysis
total_casualties = df2['nkill'] + df2['nwound']
df2['total_casualties'] = total_casualties

# Calculate the total casualties by year
casualties_by_year = df2.groupby('iyear')['total_casualties'].sum()

# Plotting the total casualties by year
plt.figure(figsize=(12, 6))
plt.plot(casualties_by_year, marker='o', color='r', linestyle='-')
plt.xlabel('Year')
plt.ylabel('Total Casualties')
plt.title('Total Casualties due to Terrorism by Year')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Analyzing the top terrorist groups and their activities
top_terrorist_groups = df2['gname'].value_counts().head(10)

# Plotting the top terrorist groups and their activities
plt.figure(figsize=(12, 6))
top_terrorist_groups.plot(kind='bar')
plt.xlabel('Terrorist Group')
plt.ylabel('Number of Incidents')
plt.title('Top Terrorist Groups and Their Activities')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Combine all terrorist group names into a single string
terrorist_groups_text = ' '.join(df2['gname'].dropna())

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(terrorist_groups_text)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud for Terrorist Group Names')
plt.show()
```

```python
# Analyzing the distribution of attacks across months
attack_month_distribution = df2['imonth'].value_counts().sort_index()

# Plotting the distribution of attacks across months
plt.figure(figsize=(12, 6))
attack_month_distribution.plot(kind='bar')
plt.xlabel('Month')
plt.ylabel('Number of Attacks')
plt.title('Distribution of Attacks Across Months')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
